Dictionary Comprehension

1. Use a dictionary comprehension to count the length of each word in a sentence.

```python
d="THis is Python this"
{word:len(word) for word in d.split()}
```

2. Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9)

```python
x={j for i in range(2,10) for j in range(1000) if j%i==0}
print(list(set(x)))
```

3. For all the numbers 1-1000, use a nested list/dictionary comprehension to find the highest single digit any of the numbers is divisible by

```python
x={k:v for v in range(2,10) for k in range(1,1000) if k%v==0}
print(x)
```
